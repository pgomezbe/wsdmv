```python
# Importing libraries
from graphviz import Digraph
import os
# Environment Variable of graphviz executable
os.environ["PATH"] += os.pathsep + 'C:/Users/pgomezbe/OneDrive - Otto Group/Escritorio/wsdmv/.venvs/windows_10_cmake_Release_Graphviz-11.0.0-win64/Graphviz-11.0.0-win64/bin/'
```

```python
# Definition of the Diagraph
dot = Digraph('AQUABUSINESSUNIT3')
dot.attr(rankdir='LR', splines='ortho')
dot.attr(labelloc='t')
dot.attr(label=r'\nData Model of AQUA Business Unit by Data Team')
dot.attr(fontsize  = '14')
dot.attr(fontcolor = 'blue')

# Metadata from Common Core (Common Business or Foundation)
metadatacc = {
    'cc_art': {
        'cohere': ('art_id')
    },
    'cc_qmauspruefprot': {
        'cohere':('id')
    },
    'cc_emailreportticket': {
        'cohere': ('ticket_id')
    }
}
# Metadata from Aqua App
metadataaqua = {
    'cc_aqualgwaeingpos': {
        'cohere':('art_id')
    },
    'cc_aqualgwaeing': {
        'cohere': ('waeingnr')
    },
    'cc_aquaqmwaeing': {
        'cohere': ('qmwaeing_id')
    },
    'cc_aquaqmwaeingpos': {
        'cohere': ('qmwaeingpos_id')
    },
    'cc_aquaqmklaerfall': {
        'cohere': ('qmklaerfall_id')
    },
    'cc_aquaqmreklamation': {
        'cohere': ('qmreklamation_id')
    },
    'cc_aqualgsendung': {
        'cohere': ('sendungsnr')
    },
    'cc_aqualgsendungpos': {
        'cohere': ('sendungsnr')
    },
    'cc_aquaqmmangel': {
        'cohere': ('qmmangel_id')
    },
    'cc_aquaqmmasstab': {
        'cohere': ('qmmasstab_id')
    },
    'cc_aquaqmnabfirma': {
        'cohere': ('qmnabfirma_id')
    },
    'cc_aquaqmpruefer': {
        'cohere': ('qmpruefer_id')
    },
    'cc_aquaqmqzahl': {
        'cohere': ('qmqzahl_id')
    },
    'cc_aquaqmtextbaustein': {
        'cohere': ('id')
    },
    'cc_aquaqmweprio': {
        'cohere': ('qmweprio_id')
    }
}
# Making nodes of subgraphs
def make_node(dot1, metadata):
    for table, info in metadata.items():
        label = f'''<<TABLE BORDER="1" CELLBORDER="1" CELLSPACING="0" CELLPADDING="4">
                    <TR><TD COLSPAN="2"><FONT POINT-SIZE="10"><B>{table}</B></FONT></TD></TR>
                    <TR><TD COLSPAN="2"><FONT POINT-SIZE="10">{info['cohere']}</FONT></TD></TR>
                </TABLE>>'''
        dot1.node(table, label=label, shape='plaintext')
    return dot1.node

# Definition of the subgraphs on the Diagraph
# NOTE: the subgraph name needs to begin with 'cluster' (all lowercase)
#       so that Graphviz recognizes it as a special cluster subgraph
with dot.subgraph(name='cluster1') as mdc:
    mdc.attr(color='#00129A')
    mdc.attr(labelloc='b')
    mdc.attr(label='METADATA COLUMNs')
    mdc.attr(fontsize='8')
    mdc.attr(fontcolor='#00129A')
    with mdc.subgraph(name='cluster2') as ship:
        ship.node('ref_12c', '''<<TABLE BORDER="1" CELLBORDER="1" CELLSPACING="0" CELLPADDING="4">
                        <TR><TD><FONT POINT-SIZE="10">wert</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">bez</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">sg</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">fr</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">ru</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">us</FONT></TD></TR>                   
                      </TABLE>>''', shape='plaintext', color="#00A2E8", fontcolor='#00A2E8')
        ship.attr(color='#00A2E8')
        ship.attr(labelloc='b')
        ship.attr(label='Shipment')
        ship.attr(fontsize  = '8')
        ship.attr(fontcolor = '#00A2E8')
    with mdc.subgraph(name='cluster3') as odwh:
        odwh.node('ref_6c', '''<<TABLE BORDER="1" CELLBORDER="1" CELLSPACING="0" CELLPADDING="4">
                        <TR><TD><FONT POINT-SIZE="10">dwh_id</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">dwh_cr_load_id</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">dwh_up_load_id</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">dwh_status</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">dwh_valid_from</FONT></TD></TR>
                        <TR><TD><FONT POINT-SIZE="10">dwh_valid_to</FONT></TD></TR>                  
                      </TABLE>>''', shape='plaintext', color="#FF0000", fontcolor='#FF0000')
        odwh.attr(color='#FF0000')
        odwh.attr(labelloc='b')
        odwh.attr(label='Oracle DWH')
        odwh.attr(fontsize  = '8')
        odwh.attr(fontcolor = '#FF0000')
with dot.subgraph(name='cluster4') as kdbuch:
    kdbuch.node('cc_kdbuch_stf_zahlplan', '''<<TABLE BORDER="1" CELLBORDER="1" CELLSPACING="0" CELLPADDING="4">
                      <TR><TD COLSPAN="2"><FONT POINT-SIZE="10">kdbuch...</FONT></TD></TR>
                      <TR><TD COLSPAN="2"><FONT POINT-SIZE="10">diadem...</FONT></TD></TR>      
                      </TABLE>>''', shape='plaintext', color='#FFC90E', fontcolor = '#FFC90E')
    kdbuch.attr(color='#FFC90E')
    kdbuch.attr(labelloc='b')
    kdbuch.attr(label='BILLING TABLES')
    kdbuch.attr(fontsize  = '8')
    kdbuch.attr(fontcolor = '#FFC90E')
with dot.subgraph(name='cluster5') as cc:
    cc.attr(color='#22B14C')
    cc.attr(labelloc='b')
    cc.attr(label='BUSINESS FOUNDATION')
    cc.attr(fontsize  = '8')
    cc.attr(fontcolor = '#22B14C')
    cc = make_node(cc, metadatacc)
with dot.subgraph(name='cluster6') as aqua:
    aqua.attr(color='#EA3FF7')
    aqua.attr(labelloc='b')
    aqua.attr(label='AQUA APP')
    aqua.attr(fontsize  = '8')
    aqua.attr(fontcolor = '#EA3FF7')
    aqua = make_node(aqua, metadataaqua)

# Relationships among created tables 
relationship = [
    ('ref_12c', 'ref_6c', ''),
    ('ref_6c', 'cc_art', ''),
    ('cc_art', 'cc_aqualgwaeingpos', ''),
    ('cc_art', 'cc_aqualgwaeing', ''),
    ('cc_art', 'cc_aquaqmwaeing', ''),
    ('cc_art', 'cc_aqualgsendungpos', ''),
    ('cc_art', 'cc_aqualgsendung', ''),
    ('ref_12c', 'cc_aqualgwaeingpos', ''),
    ('ref_12c', 'cc_aqualgwaeing', ''),
    ('ref_12c', 'cc_aqualgsendungpos', ''),
    ('ref_12c', 'cc_aqualgsendung', ''),    
    ('cc_art', 'cc_aquaqmweprio', ''),
    ('cc_aquaqmwaeing', 'cc_aquaqmklaerfall', ''),
    ('cc_aquaqmwaeing','cc_aquaqmwaeingpos', ''),
    ('ref_6c', 'cc_qmauspruefprot', ''),
    ('cc_qmauspruefprot', 'cc_aquaqmmangel', ''),
    ('cc_aquaqmklaerfall', 'cc_aquaqmmangel', ''),
    ('cc_qmauspruefprot', 'cc_aquaqmnabfirma', ''),
    ('cc_aquaqmwaeingpos', 'cc_aquaqmmasstab', ''),
    ('ref_6c', 'cc_emailreportticket', ''),
    ('cc_emailreportticket', 'cc_kdbuch_stf_zahlplan', ''),
    ('cc_emailreportticket', 'cc_aqualgsendung', ''),
    ('cc_emailreportticket', 'cc_aqualgsendungpos', ''),
    ('cc_emailreportticket', 'cc_aquaqmklaerfall', ''),
    ('cc_emailreportticket', 'cc_aquaqmreklamation', ''),
    ('cc_emailreportticket', 'cc_aquaqmmasstab', ''),
    ('cc_emailreportticket', 'cc_aquaqmpruefer', ''),
    ('cc_emailreportticket', 'cc_aquaqmqzahl', ''),
    ('cc_emailreportticket', 'cc_aquaqmtextbaustein', '')
]
# Creates these relationships iteratively
for from_table, to_table, key in relationship:
    dot.edge(from_table, to_table, headlabel=key)

dot.view()
```

```python
# Creates the PDF called as Aqua_Data_Model_v3
dot.render('Aqua_Data_Model_v3', format='pdf', view=True)
```
